```python
#Exclusion list match with COICOP All list

import pandas as pd

# Load the COICOP_all.csv file into a DataFrame
coicop_df = pd.read_csv('COICOP_all.csv')

# Define the COICOP Excl List
coicop_list = [
    "Actual rentals paid by tenants",
    "Dental services",
    "Education",
    "Electricity",
    "Financial services n.e.c",
    "FISIM1",
    "Fuels and lubricants for personal transport equipment",
    "Games of chance",
    "Gas",
    "Heat energy",
    "Hospital services",
    "Imputed rentals for owner-occupiers",
    "Insurance connected with health",
    "Insurance connected with the dwelling",
    "Insurance connected with transport",
    "Life insurance",
    "Liquid fuels",
    "Maintenance and repair of personal transport equipment",
    "Materials",
    "Medical services",
    "Motor cars",
    "Motor cycles",
    "Narcotics2",
    "Other actual rentals",
    "Other imputed rentals",
    "Other insurance",
    "Other services",
    "Paramedical services",
    "Prostitution",
    "Refuse collection",
    "Repair of household appliances",
    "Services",
    "Sewerage collection",
    "Social protection",
    "Solid fuels",
    "Therapeutic appliances and equipment",
    "Veterinary and other services for pets",
    "Water supply"
]

# List to hold the results
result = []

# Loop through the COICOP Excl List and search for each item in the DataFrame
for item in coicop_list:
    # Find the matching rows in COICOP_all.csv (adjust column names as necessary)
    matching_rows = coicop_df[coicop_df['Time period and codes'].str.contains(item, case=False, na=False)]
    
    # If matches are found, store the COICOP identifier code
    if not matching_rows.empty:
        for _, row in matching_rows.iterrows():
            result.append({'Item': item, 'COICOP Identifier Code': row['COICOP identifier code']})

# Convert result to DataFrame for better visualization
result_df = pd.DataFrame(result)

# Display the result
result_df

# Optionally, save the result to a CSV
result_df.to_csv("coicop_excl_with_codesmatch.csv", index=False)
```
