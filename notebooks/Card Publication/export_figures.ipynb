# Move datasets to bucket for review

```python
def export_files(export_path, file_name):
    export_location = f'{export_path}{file_name}'
    
    if os.path.exists(file_name) == True:

        blob = bucket.blob(export_location)
        blob.upload_from_filename(file_name) 
    
        print(f'{file_name} moved to bucket')
        
    else:
        print(f'{file_name} does not exist')
```

## Indexed spend per cardholder

```python
export_files(export_path = export_path, file_name ='Figure 1: Indexed spend per cardholder, by merchant channel.csv')
```
